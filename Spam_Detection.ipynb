# Cleaning the text data

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer 
ps = PorterStemmer()
```

```python
cleaned_data = []
for i in range(0,len(X)):
    mail = re.sub("[^a-zA-z]", " ", X[i]) # substituting any character other than a to z or A to Z
    mail = mail.lower()  # converting all the letters into small case
    mail = mail.split()  # splitting up each of the words from sentences
    # retaing the words if they are not in 'stopwords' as well as stemming all the selected words
    mail = [ps.stem(word) for word in mail if not word in set(stopwords.words('english'))] 
    mail = ' '.join(mail) # again joining those words
    cleaned_data.append(mail)
```

```python
# replacing the word 'subject' from each of the entry
cleaned = []
for j in range(0,len(cleaned_data)):
    temp = re.sub("subject", " ", cleaned_data[j])
    cleaned.append(temp)
```

# One Hot Representation & Word Embedding

```python
# Vocabulary size
vocab_size = 4500
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
onehot = [one_hot(k,vocab_size)for k in cleaned] 
```

```python
onehot
```

It is an useful tool to make all the sentences (or vectors) of equal size.

```python
# Pad Sequencing
sent_length = 30
emb_data = pad_sequences(onehot, padding = 'pre', maxlen = sent_length)
```

```python
emb_data
```

# Creating LSTM model

```python
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import Dense
from tensorflow.keras.regularizers import l2
from tensorflow.keras.models import Sequential
```

```python
vec_features = 20
model = Sequential()
model.add(Embedding(vocab_size,vec_features,input_length = sent_length))
model.add(Dropout(0.35))
model.add(LSTM(150, kernel_regularizer=l2(0.01), recurrent_regularizer=l2(0.01), bias_regularizer=l2(0.01)))
model.add(Dropout(0.35))
model.add(Dense(12, activation = 'relu', kernel_regularizer=l2(0.01), bias_regularizer=l2(0.01), activity_regularizer=l2(0.01)))
model.add(Dense(1,activation='sigmoid', kernel_regularizer=l2(0.01), bias_regularizer=l2(0.01), activity_regularizer=l2(0.01)))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model.summary())
```

Creating two numpy arrays for feeding into the LSTM model

```python
X_final = np.array(emb_data)
```

```python
y_final = np.array(y)
```

# Splitting the Dataset

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size = 0.35, random_state = 0)
```

```python
X_train, X_test
```

```python
y_train, y_test
```

```python
# Training the model
model.fit(X_train, y_train, validation_data = (X_test,y_test), epochs = 40, batch_size = 32)
```

Predicting the results

```python
y_pred  = model.predict_classes(X_test)
```

# Confusion Matrix

```python
from sklearn.metrics import confusion_matrix 
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

# Accuracy Score

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test, y_pred)
print(acc*100, "%")
```
